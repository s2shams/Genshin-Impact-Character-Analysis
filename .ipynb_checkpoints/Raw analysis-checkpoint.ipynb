```python
abyss_table = pd.read_sql("SELECT * FROM abyss", conn)
char_table = pd.read_sql("SELECT * FROM char_details", conn)
revenue_table = pd.read_sql("SELECT * FROM revenue", conn) 
```

```python
abyss_columns = abyss_table.columns.tolist()

abyss_columns
```

```python
del abyss_columns[0]
del abyss_columns[0]
del abyss_columns[-1]
```

```python
del abyss_columns[0]
abyss_columns
```

```python
for col in abyss_columns:
    abyss_table[col] = pd.to_numeric(abyss_table[col], errors='coerce')

abyss_table
```

```python
elemental_usage = abyss_table.groupby('Vision')[abyss_columns].mean()

elemental_usage
```

```python
elemental_usage['Average'] = elemental_usage.mean(axis=1, skipna=True)

elemental_usage
```

```python
overall_average = pd.read_sql(''' SELECT
        AVG(CASE WHEN Vision = 'Dendro' THEN `Average usage` END) AS Avg_dendro,
        AVG(CASE WHEN Vision = 'Cryo' THEN `Average usage` END) AS Avg_cryo,
        AVG(CASE WHEN Vision = 'Pyro' THEN `Average usage` END) AS Avg_pyro,
        AVG(CASE WHEN Vision = 'Anemo' THEN `Average usage` END) AS Avg_Anemo,
        AVG(CASE WHEN Vision = 'Electro' THEN `Average usage` END) AS Avg_Electro,
        AVG(CASE WHEN Vision = 'Hydro' THEN `Average usage` END) AS Avg_Hydro,
        AVG(CASE WHEN Vision = 'Geo' THEN `Average usage` END) AS Avg_geo
    FROM abyss;
      ''', conn)
overall_average
```

```python
five_star_average = pd.read_sql(''' SELECT
        AVG(CASE WHEN Vision = 'Dendro' THEN `Average usage` END) AS Avg_dendro,
        AVG(CASE WHEN Vision = 'Cryo' THEN `Average usage` END) AS Avg_cryo,
        AVG(CASE WHEN Vision = 'Pyro' THEN `Average usage` END) AS Avg_pyro,
        AVG(CASE WHEN Vision = 'Anemo' THEN `Average usage` END) AS Avg_Anemo,
        AVG(CASE WHEN Vision = 'Electro' THEN `Average usage` END) AS Avg_Electro,
        AVG(CASE WHEN Vision = 'Hydro' THEN `Average usage` END) AS Avg_Hydro,
        AVG(CASE WHEN Vision = 'Geo' THEN `Average usage` END) AS Avg_geo
    FROM abyss
    WHERE Rarity = 5;
      ''', conn)

five_star_average
```

```python
four_star_average = pd.read_sql(''' SELECT
        AVG(CASE WHEN Vision = 'Dendro' THEN `Average usage` END) AS Avg_dendro,
        AVG(CASE WHEN Vision = 'Cryo' THEN `Average usage` END) AS Avg_cryo,
        AVG(CASE WHEN Vision = 'Pyro' THEN `Average usage` END) AS Avg_pyro,
        AVG(CASE WHEN Vision = 'Anemo' THEN `Average usage` END) AS Avg_Anemo,
        AVG(CASE WHEN Vision = 'Electro' THEN `Average usage` END) AS Avg_Electro,
        AVG(CASE WHEN Vision = 'Hydro' THEN `Average usage` END) AS Avg_Hydro,
        AVG(CASE WHEN Vision = 'Geo' THEN `Average usage` END) AS Avg_geo
    FROM abyss
    WHERE Rarity = 4;
      ''', conn)
four_star_average
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))

elements = ['Dendro', 'Cryo', 'Pyro', 'Anemo', 'Electro', 'Hydro', 'Geo']
colors = ['green', 'lightblue', 'red', 'aquamarine', 'purple', 'blue', 'gold']
plt.bar(elements, overall_average.iloc[0].tolist(), color=colors)

plt.xlabel('Element')
plt.ylabel('Usage rate')
plt.title('Spiral abyss usage rate')

plt.show()
```

```python
def create_bar(x, label):
    plt.figure(figsize=(8, 6))

    elements = ['Dendro', 'Cryo', 'Pyro', 'Anemo', 'Electro', 'Hydro', 'Geo']
    colors = ['green', 'lightblue', 'red', 'aquamarine', 'purple', 'blue', 'goldenrod']
    plt.bar(elements, x.iloc[0].tolist(), color=colors)

    plt.ylabel('Usage rate')
    plt.grid(True)
    plt.title(f'{label} Spiral Abyss Usage Rate')

    plt.show()
```

```python
create_bar(overall_average, 'Overall')
create_bar(five_star_average, '5 star')
create_bar(four_star_average, '4 star')
```

```python
elemental_usage['Vision'] = ['Anemo', 'Cryo', 'Dendro', 'Electro', 'Geo', 'Hydro', 'Pyro']
del elemental_usage['Average']

version_usage = pd.melt(elemental_usage, id_vars=['Vision'], var_name='Version', value_name='Usage')
```

```python
elemental_usage
```

```python
plt.figure(figsize=(13, 6))

colors = ['aquamarine', 'lightblue', 'green', 'purple', 'goldenrod', 'blue', 'red']

for idx, element in enumerate(version_usage["Vision"].unique()):
    element_subset = version_usage[version_usage["Vision"] == element]
    plt.plot(
        element_subset["Version"],
        element_subset["Usage"],
        label=element,
        color=colors[idx % len(colors)]  
    )

plt.xlabel("Version")
plt.ylabel("Usage")
plt.title("Elemental Usage rate")
plt.legend()
plt.grid(True)
plt.show()
```
