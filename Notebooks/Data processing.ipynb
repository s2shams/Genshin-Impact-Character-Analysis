# Abyss Table

#### The first issue I notice is that the data in the abyss columns aren't numeric. I fix this by first creating a list of the columns I wan't to turn into the numeric type and use pandas to convert them.

```python
abyss_cols = abyss.columns.tolist()
del abyss_cols[0]
del abyss_cols[0]
del abyss_cols[0]
del abyss_cols[-1]
abyss_cols
```

```python
for col in abyss_cols:
    abyss[col] = pd.to_numeric(abyss[col], errors='coerce')
    
abyss
```

```python
abyss.info()
```

#### Now we get rid of the null values, considering the table consists of usage rate, if it's null, it indicates the character was not used at all during the version patch. Now while this may seem to extreme, it simply means that the character has not been released at the time of patch.

```python
abyss = abyss.replace(np.nan, '', regex=True)
abyss
```

```python
abyss.info()
```

#### The issue with this is that it changes the type back into objects, which is not what I want. I'll revert the changes and keep the null values. This will not be an issue since the visualization tools I'll be using (Power BI and sns/matplotlib) can easily just ignore the null values.

```python
for col in abyss_cols:
    abyss[col] = pd.to_numeric(abyss[col], errors='coerce')
    
abyss
```

# Elemental Usage

#### Next I want to create a table that'll allow me to see the elemental usage rate over time. I do this by using the grouping function to group together the elements throughout their versions.

```python
elemental_usage = abyss.groupby('Vision')[abyss_cols].mean()

elemental_usage
```

#### It looks a bit hard to read so I limited it to 2 decimal places. Note that we leave the null values in due to the same reason above.

```python
elemental_usage = elemental_usage.round(decimals=2)
elemental_usage
```

# Breaking Down The Double Banners

#### The first step is to split the column. 1 column for each character, if a character is not present, the 2nd column will simply contain 'none'.

```python
revenue_1 = revenue['Banner (IOS CN Market only)'].str.split('&', n=1, expand=True)
```

```python
revenue_1.columns = ['character1', 'character2']
revenue = revenue.drop(['Banner (IOS CN Market only)'], axis=1)
revenue = pd.concat([revenue, revenue_1], axis=1)
```

#### Next, I repeat the same step except for the elements

```python
revenue_1 = revenue['Element(s)'].str.split(',', n=1, expand=True)
revenue_1.columns = ['element1', 'element2']
revenue = revenue.drop(['Element(s)'], axis=1)
revenue = pd.concat([revenue, revenue_1], axis=1)
```

#### Now I want to create 2 additional columns, each for the corresponding character. The column will indicate whether it is a rerun banner and if it is, then which rerun it is.

```python
def find_index(x):
    for index, letter in enumerate(x):
        if letter ==  '(':
            return index

rerun1 = []        
for char in revenue['character1']:
    if '(' in char:
        index = find_index(char);
        rerun1.append(char[index:])
    else:
        rerun1.append('no')
revenue['rerun1'] = pd.DataFrame(rerun1)
```

```python
revenue['character2'].fillna('none', inplace=True)
revenue['element2'].fillna('none', inplace=True)

temp = []
for char in revenue['character2']:
    if '(' in char:
        index = find_index(char)
        temp.append(char[index:])
    else:
        temp.append('no')

rerun1_v2 = []
rerun2 = []

for char in temp:
    if '&' in char:
        rerun1_v2.append(char.split('&')[0])
        rerun2.append(char.split('&')[1])
    else:
        rerun1_v2.append('no')
        rerun2.append(char)
revenue['rerun2'] = pd.DataFrame(rerun2)
```

```python
for index, char in enumerate(revenue['rerun1']):
    if char != rerun1_v2[index]:
        revenue['rerun1'][index] = rerun1_v2[index]
        
for index,char in enumerate(revenue['character1']):
    if '(Rerun)' in char:
        revenue['rerun1'][index] = '1'

for index, char in enumerate(revenue['character2']):
    if '(Rerun)' in char:
        revenue['rerun2'][index] = '1'
revenue
```

#### The columns have been successfully created. Now I'll clean up the columns to make it easier to analyse

```python
revenue['character1'] = revenue.character1.apply(lambda x: x.split('(')[0] if '(' in x else x)
revenue['character2'] = revenue.character2.apply(lambda x: x.split('(')[0] if '(' in x else x)
revenue['rerun1'] = revenue['rerun1'].map(lambda i: ''.join([x for x in i if x.isdigit()]))
revenue['rerun2'] = revenue['rerun2'].map(lambda i: ''.join([x for x in i if x.isdigit()]))
revenue.rename(columns={'rerun1':'banner1', 'rerun2':'banner2'}, inplace=True)
```

#### I decided it'd be easier to work with the banner number (i.e is it the first, second or third banner) instead of seeing whether it's a rerun or not

```python
revenue['banner1'][46] = '2'
revenue['banner1'] = pd.to_numeric(revenue['banner1'])
revenue['banner2'] = pd.to_numeric(revenue['banner2'])
revenue['banner1'].fillna(0, inplace=True)
revenue['banner2'].fillna(0, inplace=True)

revenue['banner1'] = revenue['banner1'] + 1
revenue['banner2'] = revenue['banner2'] + 1

for index, char in enumerate(revenue['character2']):
    if char == 'none':
        revenue['banner2'][index] = 0

revenue
```
